```python
import pandas as pd

data = pd.read_csv('/content/Mini Project - Marks Adding.csv')
data.head()
```

```python
# Group the data by roll number and sum the marks
unique_roll_numbers = data.groupby('Roll Num')['Marks'].sum()

# Print the unique roll numbers and their total marks
print(unique_roll_numbers)
```
